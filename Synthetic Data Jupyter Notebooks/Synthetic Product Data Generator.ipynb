```python
import csv
import random
from faker import Faker

# Initialize Faker
fake = Faker()

# Sample categories and sub-categories
categories = {
    "Electronics": ["Computers", "Mobile Devices", "Accessories","Audio & Entertainment", "Visual Electronics", "Gaming"],
    "Fashion": ["Men's Clothing", "Women's Clothing", "Shoes", "Bags", "Activewear", "Luxury Fashion"],
    "Health & Beauty": ["Skincare", "Haircare", "Wellness","Makeup", "Fragrances", "Men’s Grooming"],
    "Home & Kitchen": ["Furniture", "Appliances", "Decor","Cookware & Bakeware", "Tableware", "Lighting"],
    "Sports & Outdoors": ["Fitness", "Camping", "Cycling","Water Sports", "Running & Walking", "Outdoor Recreation"],
}

product_types = {
    # Electronics
    "Computers": ["Laptop", "Desktop", "Tablet", "Gaming PC", "Workstation"],
    "Mobile Devices": ["Smartphone", "Smartwatch", "Foldable Phone", "E-Reader"],
    "Accessories": ["Wireless Mouse", "USB-C Hub", "Bluetooth Earbuds", "External Hard Drive"],
    "Audio & Entertainment": ["Headphones", "Soundbar", "Home Theater System", "Podcast Microphone"],
    "Visual Electronics": ["Monitor", "Projector", "Smart TV", "DSLR Camera"],
    "Gaming": ["Gaming Mouse", "Mechanical Keyboard", "Handheld Console", "VR Headset"],

    # Fashion
    "Men's Clothing": ["T-Shirt", "Jeans", "Jacket", "Dress Shirt"],
    "Women's Clothing": ["Dress", "Blouse", "Sweater", "Leggings"],
    "Shoes": ["Sneakers", "Boots", "Sandals", "Loafers"],
    "Bags": ["Backpack", "Tote Bag", "Messenger Bag", "Clutch"],
    "Activewear": ["Running Shorts", "Compression Shirt", "Yoga Pants", "Athletic Jacket"],
    "Luxury Fashion": ["Designer Handbag", "Leather Jacket", "Silk Scarf", "Tailored Suit"],

    # Health & Beauty
    "Skincare": ["Face Serum", "Moisturizer", "Sunscreen", "Clay Mask"],
    "Haircare": ["Shampoo", "Conditioner", "Hair Gel", "Curling Iron"],
    "Wellness": ["Vitamin Supplements", "Essential Oils", "Massage Roller", "Aromatherapy Diffuser"],
    "Makeup": ["Foundation", "Lipstick", "Eyeliner", "Mascara"],
    "Fragrances": ["Perfume", "Cologne", "Body Spray", "Roll-on Oil"],
    "Men’s Grooming": ["Beard Oil", "Shaving Kit", "Aftershave", "Hair Pomade"],

    # Home & Kitchen
    "Furniture": ["Sofa", "Coffee Table", "Bookshelf", "Dining Chair"],
    "Appliances": ["Microwave", "Blender", "Washing Machine", "Vacuum Cleaner"],
    "Decor": ["Wall Art", "Floor Lamp", "Throw Pillow", "Candles"],
    "Cookware & Bakeware": ["Nonstick Pan", "Baking Sheet", "Mixing Bowl", "Cutting Board"],
    "Tableware": ["Dinner Plate", "Wine Glass", "Flatware Set", "Serving Platter"],
    "Lighting": ["Ceiling Lamp", "LED Strip Lights", "Desk Lamp", "Outdoor Lantern"],

    # Sports & Outdoors
    "Fitness": ["Dumbbells", "Jump Rope", "Treadmill", "Resistance Bands"],
    "Camping": ["Sleeping Bag", "Tent", "Portable Grill", "Camp Stove"],
    "Cycling": ["Road Bike", "Mountain Bike", "Folding Bicycle", "Bike Helmet", "Bike Lock"],
    "Water Sports": ["Kayak", "Paddleboard", "Snorkel Set", "Swim Goggles"],
    "Running & Walking": ["Running Shoes", "Smart Fitness Tracker", "Hydration Pack", "Compression Socks"],
    "Outdoor Recreation": ["Binoculars", "Trekking Poles", "Hiking Backpack", "GPS Tracker"]
}

# Generate synthetic data
num_products = 8999  # Number of products you want
data = [["Product ID","Product Name", "Category", "Sub-Category","Brand"]]


adjectives = ["Ultra", "Pro", "Max", "Lite", "Premium", "Classic", "Essential", "Eco", "Smart"]
id = 1000

for _ in range(num_products):

    category = random.choice(list(categories.keys()))
    sub_category = random.choice(categories[category])
    
    # Ensure product type aligns with the correct sub-category
    product_type = random.choice(product_types.get(sub_category, [sub_category]))  # Fallback to sub-category
    
    adjective = random.choice(adjectives + [""])  # Allow blanks for variety
    brand = fake.company()
    
    # Ensure name logic matches its correct category/sub-category
    product_name = f"{adjective} {product_type}".strip()
    
    #product_name = f"{brand} {adjective} {product_type}".strip()
    #product_name = f"{brand} {adjective} {sub_category}" 
    data.append([id, product_name, category, sub_category, brand])
    id += 1

# Write to CSV
filename = "synthetic_inventory.csv"
with open(filename, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print(f"{num_products} products saved to {filename}!")
```
