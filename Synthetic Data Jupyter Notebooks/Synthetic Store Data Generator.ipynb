```python
import csv
import random
from faker import Faker

fake = Faker()

regions = {
    "Northeast": ["New York, NY", "Boston, MA", "Philadelphia, PA", "Providence, RI", "Buffalo, NY", "Hartford, CT"],
    "Midwest": ["Chicago, IL", "Columbus, OH", "Detroit, MI", "Minneapolis, MN", "St. Louis, MO"],
    "South": ["Atlanta, GA", "Dallas, TX", "Miami, FL", "Charlotte, NC", "Houston, TX", "Nashville, TN", "Fort Worth, TX", "Orlando, FL"],
    "West": ["Los Angeles, CA", "Seattle, WA", "Denver, CO", "San Francisco, CA", "Phoenix, AZ", "San Diego, CA", "Las Vegas, NV", "Portland, OR"]
}

    # Midwest
    "Chicago, IL": ["Lincoln Park", "Hyde Park", "Wicker Park", "Brighton Park", "Englewood", "Lakeview", "Evanston", "Oak Park", "Avalon Park", "West Loop"],
    "Columbus, OH": ["German Village", "Short North", "Clintonville", "Bexley", "Franklinton", "Dublin", "Arena District"],
    "Detroit, MI": ["Dearborn", "Greektown", "Delray", "Franklin", "Ferndale", "Southfield", "Grosse Pointe", "Warren", "Troy", ],
    "Minneapolis, MN": ["Edina", "Bloomington", "St. Louis Park", "Roseville", "Maple Grove", "Eagan", "Woodbury"],
    "Indianapolis, IN": ["Broad Ripple", "Fountain Square", "Meridian-Kessler", "Lawrence", "Castleton", "Wayne", "Eastside"],
    "St. Louis, MO": ["Central West End", "Soulard", "The Hill", "Downtown", "Tower Grove", "Lafayette Square", "Maplewood"],
    
    # South
    "Atlanta, GA": ["Five Points", "South Downtown", "Ansley Park", "Home Park", "Berkeley Park", "Marietta", "High Point", "Summerhill"],
    "Dallas, TX": ["Denton", "Forney", "Waxahachie", "Flower Mound", "Frisco", "Southlake", "University Park" "Plano", "Wylie", "Allen", "Richardson",  "Desoto", "Cedar Hill", "Mansfield"],
    "Fort Worth TX": ["Ederville", "Meadowbrook", "Northside", "River Oaks", "Oakhurst", "Springdale", "Alamo Heights"], 
    "Miami, FL": ["Coral Gables", "Little Havana", "South Beach", "Wynwood", "Doral", "Kendall", "Miami Springs"],
    "Charlotte, NC": ["Ballantyne", "SouthPark", "NoDa", "University City", "Dilworth", "Plaza Midwood"],
    "Houston, TX": ["The Woodlands", "Sugar Land", "Katy", "Pearland", "Cypress", "Spring"],
    "Nashville, TN": ["East Nashville", "Belle Meade", "Green Hill", "Downtown", "Sylvan Park", "Germantown", "Inglewood", "South Nashville"],
    "Orlando, FL": ["Azalea Park", "Bel Air", "Callahan", "South Eola", "Conway", "Randal Park", "Vista Park", "Eaton Ville", "Kirkman", "Winter Park", "Winter Garden",  "Kissimmee", "Altamonte Springs", "Oviedo", "Sanford"],
    
    # West
    "Los Angeles, CA": ["Hollywood", "Beverly Hills", "Pasadena", "Santa Monica", "Long Beach", "Glendale"],
    "Seattle, WA": ["Ballard", "Capitol Hill", "Fremont", "Queen Anne", "West Seattle", "Greenwood"],
    "Denver, CO": ["Cherry Creek", "Highlands", "Five Points", "Capitol Hill", "Aurora", "Lakewood"],
    "San Francisco, CA": ["Mission District", "Pacific Heights", "Noe Valley", "Castro", "Sunset District", "Marina"],
    "Phoenix, AZ": ["Scottsdale", "Tempe", "Mesa", "Chandler", "Gilbert", "Glendale"],
    "San Diego, CA": ["La Jolla", "Pacific Beach", "Mission Valley", "Chula Vista", "Encinitas", "Carlsbad"],
    "Las Vegas, NV": ["Henderson", "Summerlin", "North Las Vegas", "Paradise", "Spring Valley", "Sunrise Manor", "Enterprise", "Boulder City"],
    "Portland, OR": ["Beaverton", "Gresham", "Hillsboro", "Lake Oswego", "Tigard", "Milwaukie"]

store_name_count = {}  # Track occurrences of store names
num_stores = 99
max_store_count_per_suburb = 2  # Set threshold to change city

data = [["Store ID", "Store Name", "City", "State", "Region"]]

for store_id in range(1, num_stores + 1):
    retry_count = 0
    
    while retry_count < 20:  # Prevent infinite loops
        region = random.choice(list(regions.keys()))
        city_state = random.choice(regions[region])
        city, state = city_state.split(", ")
        
        # Select valid suburb or fallback
        if city_state in suburbs:
            suburb = random.choice(suburbs[city_state]).title()
        else:
            suburb = fake.city_suffix()  # Fallback

        base_store_name = f"{suburb}, {city} Store"

        # If store count exceeds limit, pick a new city
        if store_name_count.get(base_store_name, 0) < max_store_count_per_suburb:
            store_name_count[base_store_name] = store_name_count.get(base_store_name, 0) + 1
            store_name = f"{base_store_name} {store_name_count[base_store_name]}" if store_name_count[base_store_name] > 1 else base_store_name
            break  # Stop retrying once we find a valid store name
        else:
            retry_count += 1  # Try a new city

    data.append([store_id, store_name.title(), city, state, region])

# Write to CSV
filename = "synthetic_store_data.csv"
with open(filename, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)

print(f"{num_stores} stores saved to {filename}!")
```
